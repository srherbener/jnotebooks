# Notes, 12/29/17
* Decided to use Sphinx/ReadTheDocs instead of GitHub Pages since Sphinx/ReadTheDocs is independent of GitHub
* Tutorial
    * http://dont-be-afraid-to-commit.readthedocs.io/en/latest/documentation.html#documentation-using-sphinx-and-readthedocs-org
    * create anaconda environment for documentation
        * conda create -n docs python=3.6 sphinx
        * source activate docs
    * create a git repository to hold the documentation (following directions on above website)
        * mkdir ~/Documents/Sphinx/my-first-docs
        * cd ~/Documents/Sphinx/my-first-docs
        * git init
        * git remote add origin git@github.com:srherbener/my-first-docs.git
        * git checkout -b master
        * mkdir docs
        * sphinx-quickstart
            * root directory --> docs
            * separate source and build directories --> y
            * name prefix for templates and static dirs --> hit return to leave as underscore
            * project name --> Steve's first docs
            * project id --> 1.0
            * project release --> 1.0.1
            * project language --> hit return to leave as english (en)
            * source file suffix --> hit return to leave as .rst
            * name of your master document --> hit return to leave as "index"
            * want to use epub builder --> n
            * say yes to all sphinx extensions except for imgmath and githubpages
            * create makefile --> y
            * create window command file --> n
        * cd docs
        * edit source/index.rst
    * update github repository
        * git commit -a
        * git push origin master

* Lightweight markup languages  
    https://en.wikipedia.org/wiki/Lightweight_markup_language
    
    * reStructuredText and Markdown seem to be most popular
    * Both support inline LaTex for creating nice formatted equations
    * Both are used by popular document creation tools
    * Markdown
        * Pros: simplest language, easiest to learn
        * Cons: no standard, many variants
    * reStructuredText
        * Pros: 
            * good tradeoff between simple and easy to use versus flexibility
            * there is a standard
                * new features are added via extensions (of which there are many)
        * Cons: hard to find any, but definitely not as easy to use as Markdown
* Documentation creation tools  
    https://en.wikipedia.org/wiki/Comparison_of_documentation_generators
    
    * Doxygen  
        http://www.stack.nl/~dimitri/doxygen/
        
        * Source is embedded in C++ code
        * Can create html, man pages and API references
        * Not clear how to get to 
    * Jekyll  
        https://jekyllrb.com/
        
        * Used by github pages
            * Can be shut off if guthub pages is given a ready to use format such as html
        * Reads Markdown, Textile or Liquid
        * Ruby based
        * Can create html, but cannot create man pages
    * Sphinx  
        http://www.sphinx-doc.org/en/stable/index.html
        
        * Reads reStructuredText
        * Python based
        * Can create html and man pages
        * Can create API references for Python, not clear if can do this for C++
* Documentation publishing
    * GitHub Pages  
        https://pages.github.com/
        
        * Markdown is default, uses Jekyll to build html
        * Can shut off Jekyll and read html directly
    * ReadTheDocs  
        https://readthedocs.org/
        
        * reStructuredText, uses Sphinx to build html
            * Not clear if can process html directly

    * There are paths from Sphinx to GitHub Pages  
        https://daler.github.io/sphinxdoc-test/
        
        * Complicated to set up, but not so bad to use (after set up)
        * Path
            * Create reStructuredText source in /docs of branch master
            * Use Sphinx to create html
            * ReadTheDocs picks up commits on branch master
            * Copy html into / of branch gh-pages
                * First time: create an empty file (touch) called ".nojekyll"
                * This tells Jekyll (GitHub Pages) that you are supplying HTML directly
            * GitHub Pages pickup up commits on branch gh-pages  
            
            * Pros: one source format
            * Cons: checking generated files into gh-pages branch
        * Perhaps GitHub Pages can process reStructuredText directly?
