## Getting started doc
* https://www.tensorflow.org/get_started/get_started
* Multiple APIs
    * TensorFlow Core
        * Low level
        * High access, control
        * Good for developers
    * Higher level APIs
        * Good for consistency
        * Good for users
* Tensor
    * scalar value --> tensor, rank 0
    * vector --> tensor, rank 1
    * matrix --> tensor, rank 2
    * 3D array --> tensor, rank 3
    * etc.
* Two sections to TensorFlow
    1. Build a computational graph
        * Create nodes which hold values
        * Tie nodes together with edges which perform computations
    1. Run the computational graph
        * Create a session object
        * Select a node from the graph
        * Execute the run method of the session object
* TensorBoard
    * Utility to draw a picture of the computational graph
* Optimization
    * train object contains optimization methods (e.g., gradient descent)
    * create a model (computational graph)
        * variable objects are the parameters of the model that will get optimized
        * placeholder objects are the independent and dependent variables of the model
    * create a cost function
    * set the dependent variables to desired values
    * set the independent variables to initial values
    * run the optimize method
        * modifies the parameters and evaluates using the cost function
            * compare model output using independent variables with desired values (cost function)
            * modify independent variables
            * repeat until cost function is minimized/maximized
        * returns the optimized parameter values
* Estimator API
    * Higher level than train API
        * Easier to organize a machine learning session
    * Lots of built in models
    * Can also use a custom model

## Install TensorFlow on Mac OS
* Install in annaconda
    * conda create -n tensorflow python=3.5
        * this creates a conda environment (container) that is based on python 3.5
        * Use 3.5 instead of 3.6 since tensorflow issues a warning that 3.5 is highest version of python3 that it supports
    * source activate tensorflow
        * Switch to the new tensorflow environment
    * pip install --ignore-installed --upgrade https://storage.googleapis.com/tensorflow/mac/cpu/tensorflow-1.4.0-py3-none-any.whl
        * This installs tensorflow for python3

* Validate the installation
    * source activate tensorflow
    * Run the following code:
    
        import tensorflow as tf  
        hello = tf.constant('Hello, TensorFlow!')  
        sess = tf.Session()  
        print(sess.run(hello))  

    * When I run this, I get a warning:
    
        2017-12-29 10:04:49.232092: I tensorflow/core/platform/cpu_feature_guard.cc:137] Your CPU supports instructions that this TensorFlow binary was not compiled to use: SSE4.1 SSE4.2 AVX  
        
    * This shows up right after the sess = ts.Session() command
    * The warning can be surppressed by (but this suppresses all warnings):
    
        export TF_CPP_MIN_LOG_LEVEL=2.

        TF_CPP_MIN_LOG_LEVEL is a TensorFlow environment variable responsible for the logs, to silence INFO logs set it to 1, to filter out WARNING 2 and to additionally silence ERROR logs (not recommended) set it to 3
